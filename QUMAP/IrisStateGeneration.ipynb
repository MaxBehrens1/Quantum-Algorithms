# Imports and constructs the states for the Iris dataset

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
iris = fetch_ucirepo(id=53) 
  
# Import data
iris_df = iris.data.features
iris_df = 2 * (iris_df - iris_df.min()) / (iris_df.max() - iris_df.min()) - 1

targets = iris.data.targets

# Join the tables
iris_df['plant_type'] = targets

# Save the DataFrame to a CSV file
iris_df.to_csv('iris_dataset.csv', index=False)
```
